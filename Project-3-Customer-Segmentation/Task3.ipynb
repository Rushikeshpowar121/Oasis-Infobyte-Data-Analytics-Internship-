```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ifood_df.csv")
```

```python
df.head()
```

## Feature for Segmentation

```python
X = df[['Income', 'MntTotal']]
```

## Apply K-Means clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Customer_Segment'] = kmeans.fit_predict(X)
```

## Visualize customer segments

```python
plt.scatter(df['Income'], df['MntTotal'])
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.title("Customer Segmentation")
plt.show()
```

```python
df.groupby('Customer_Segment')[['Income','MntTotal']].mean()
```

```python
df.to_csv("customer_segmentation_result.csv", index=False)
```

## Customer Segmentation Analysis

Customer data was analyzed to identify distinct groups based on income and total spending behavior. 
K-Means clustering was applied to segment customers into three groups with similar purchasing patterns.

The analysis revealed variations in spending behavior across income levels, 
providing insights for targeted marketing strategies and customer relationship management.


## Business Recommendations

1. High-income, high-spending customers should be targeted with premium offers and loyalty programs.

2. Moderate customers can be encouraged to increase purchases through personalized promotions.

3. Low-spending customers may respond to discounts and engagement campaigns.

4. Marketing strategies should be customized based on customer segment behavior.
