```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Loading And Inspenction

```python
df = pd.read_csv("menu.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning

```python
# Check the Missing Values
df.isnull().sum()
```

## Descriptive Statistics

```python
# Average nutrition values
df[['Calories','Total Fat','Carbohydrates','Protein']].mean()
```

```python
# Highest calorie item
df.loc[df['Calories'].idxmax()]
```

## Category Analysis (Retail Insight)

```python
#Category with heighest Calories
df.groupby('Category')['Calories'].mean()
```

```python
# Item Per Category 
df['Category'].value_counts()
```

## Data Vsualization

```python
# Calories By Category 
df.groupby('Category')['Calories'].mean().plot(kind='bar')
plt.title("Average Calories by Category")
plt.show()
```

```python
# Calories VS Protein
plt.scatter(df['Calories'], df['Protein'])
plt.xlabel("Calories")
plt.ylabel("Protein")
plt.show()
```

## Business Recommendation

Based on the exploratory data analysis of the retail menu dataset, the following recommendations are suggested:

### 1. Introduce more low-calorie menu options
Several items, especially in the breakfast category, have high calorie values. Adding healthier alternatives can attract health-conscious customers.

### 2. Promote high-protein items
Products with higher protein content can be marketed toward fitness-focused consumers to increase sales in that segment.

### 3. Highlight nutritional information clearly
Displaying calorie and fat content prominently can help customers make informed decisions and build brand trust.

### 4. Develop targeted marketing by category
Categories with higher average calories or popularity can be promoted strategically to maximize revenue.

### 5. Offer balanced meal combinations
Combine high-calorie items with low-calorie beverages or sides to provide healthier meal choices.
