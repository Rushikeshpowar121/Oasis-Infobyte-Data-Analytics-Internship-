```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open("CA_category_id.json") as f:
    data = json.load(f)

df = pd.json_normalize(data)
df.head()
```

```python
items_df = pd.json_normalize(df['items'])
items_df.head()
```

```python
df_clean = pd.concat([df.drop(columns=['items']), items_df], axis=1)
df_clean.head()
```

```python
# Flatten nested column
tems_df = pd.json_normalize(df['items'])
df_clean = pd.concat([df.drop(columns=['items']), items_df], axis=1)
df_clean.head()
```

## Data Cleaning 

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.astype(str)
```

```python
df_clean = df_clean.drop_duplicates()
df_clean.duplicated().sum()
```

```python
df_clean.columns = df_clean.columns.str.replace(".", "_")
df_clean.columns = df_clean.columns.str.lower()
```

```python
df_clean = df_clean.astype(str)
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.info()
df_clean.head()
```

```python
df_clean.to_csv("cleaned_dataset.csv", index=False)
```
